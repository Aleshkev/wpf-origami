```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv("_build/default/g1.csv").pivot("y", "x", "v")
data
```

```python
fig, ax = plt.subplots(figsize=(15,15))
ax.imshow(data, interpolation='nearest')
xticks = list(zip(*[(i, v) for i, v in enumerate(data.axes[0]) if round(v) == v]))
ax.set_xticks(xticks[0])
ax.set_xticklabels(xticks[1])
yticks = list(zip(*[(i, v) for i, v in enumerate(data.axes[1]) if round(v) == v]))
ax.set_yticks(yticks[0])
ax.set_yticklabels(yticks[1])
# ax.set_yticklabels(data.axes[1])
plt.show()
```
